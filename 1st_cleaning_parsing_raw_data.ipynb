```python
import json
with open("finaldata.txt",encoding="utf-8") as f:   #encoding is used because it cannot read some data so for this it is used
    data=f.read()

    
    
```

```python
#FIRST WE MAKE A CHUNK ONE CHUNK IS ONE USER AND WE HAVE STORED ALL THIIR INFO IN A DIC OF ALL USERS
chunks=data.split("\n\n")
chunks=[c for c in chunks if c.strip()]#the minimum length of the chunk should be 5 otherwise it is not a chunk and don't include it
# chunks = [c for c in chunks if len(c.strip().split("\n")) >= 5] this can be used if we want that our every user must contain 5 fields
def parsing_chunks(chunk):
    try:
        raw_chunk=chunk.strip()
        new_chunk=raw_chunk.split("\n")
        
        user_name=new_chunk[0]
        posts=int(new_chunk[1].split()[0].replace(",",""))    #we are removing the posts string from the post
        followers=new_chunk[2].split()[0].replace(",","")
        if "K" in followers:
            updated_followers=float(followers.replace("K",""))
            no_of_followers=int(updated_followers*1000)
        else:
            no_of_followers=int(followers)
                      
        followings=new_chunk[3].split()[0].replace(",","")
        if "K" in followings:
            updated_followings=float(followings.replace("K",""))
            no_of_followings=int(updated_followings*1000)
        else:
            no_of_followings=int(followings)
        
        name=new_chunk[4]
        if len(new_chunk)>5:
            page_type=new_chunk[5]
            bio=" ".join(new_chunk[6:])   # if we directly assign by slicing it will create the list
        else:
            page_type="unknown"
            bio=""
            
        
        
        return {"user_name":user_name,
                "posts":posts,
                "followers":no_of_followers,
                "followings":no_of_followings,
                "name":name,
                "page_type":page_type,
                "bio": bio }
    except Exception as error_name:
        print(error_name,chunk)    
    
```

```python

#NOW WE ARE GOING TO MAKE A list IN WHICH ALL THE USERS INFO IS STORED IN A FORM OF INDIVITUAL DICTIONARY
all_users_info=[]
for chunk in chunks:
    user_info=parsing_chunks(chunk)
    if user_info:
        all_users_info.append(user_info)
print(all_users_info)

with open("final_users_data.json","w") as f:
    json.dump(all_users_info,f)

```
